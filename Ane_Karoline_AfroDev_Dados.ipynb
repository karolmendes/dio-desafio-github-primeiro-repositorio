```python
#Importação das Bibliotecas e Documentos as Serem Analisados

import pandas as pd
import numpy as np
from pandas import read_excel
pd.set_option("display.max_rows", 100)

df1 = pd.read_excel("/content/drive/MyDrive/Athletes.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Coaches.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/EntriesGender.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Medals.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Teams.xlsx")
```
